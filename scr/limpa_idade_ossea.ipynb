```python
def limpa_texto(texto: str):
    texto = remove_string(texto, '*', '')
    texto = remove_string(texto, '..', '.')
    texto = remove_string(texto, '  ', ' ')
    texto = remove_string(texto, ' .', '.')
    texto = remove_string(texto, ' ,', ',')
    texto = remove_string(texto, '( ', '(')
    texto = remove_string(texto, ' )', ')')
    texto = remove_string(texto, ' :', ':')
    texto = remove_string(texto, '\\n', '\n')
    texto = remove_string(texto, '\t', '')
    texto = remove_string(texto, '\n ', '\n')
    texto = remove_string(texto, '\n.', '\n')
    texto = remove_string(texto, '\n\n', '\n')
    texto = texto.strip()
    texto = texto.strip('\n')
    
    return texto
```

```python
def remove_string(texto: str, procurado, substituido: str):
    if type(procurado) is str:
        while procurado in texto:
            texto = texto.replace(procurado, substituido)
        return texto
    elif type(procurado) is list:
        for palavra in procurado:
            while palavra in texto:
                texto = texto.replace(palavra, substituido)
        return texto
```

```python
def lista_frases(texto: str, lista_frases_procuradas: list):
        
    lista_return = []
    frases_texto = texto.split('\n')
    
    for frase in frases_texto:
        for procurada in lista_frases_procuradas:
            if procurada.upper() in frase.upper():
                lista_return.append(frase)
                break
                
    #lista_return = ' '.join(lista_return)
    return lista_return
```

```python
def remove_frases(texto: str, lista_frases_procuradas: list):
    
    lista_return = []
    texto = texto.replace('Tecnica:', '')
    texto = texto.replace('. ', '\n')
    texto = texto.split('\n')
    #texto = texto.split('.')
    
    for frase in texto:
        flag_remover = False
        for procurada in lista_frases_procuradas:
            if procurada.upper() in frase.upper():
                flag_remover = True
                break
        
        if not flag_remover:
            lista_return.append(frase)
    
    lista_return = '.\n'.join(lista_return)
    lista_return = limpa_texto(lista_return)
    
    return lista_return
```

```python
def contem_frase(texto: str, lista_frases_procuradas: list):
    for procurada in lista_frases_procuradas:
        if procurada.upper() in texto.upper():
            return True
                
    return False
```

```python
def remove_stopwords(texto: str):
    
    if type(texto) is str:
        lista_texto = [texto]
    elif type(texto) is list:
        lista_texto = texto
    else:
        print("A função remove_stopwords aceita como parâmetro apenas strings ou listas.")
        exit()
    
    lista_removida = []        
    for text in lista_texto:
        text_tokens = nltk.tokenize.word_tokenize(text, language='portuguese') 
        tokens_without_sw = [word for word in text_tokens if not word in stopwords]
        tokens_without_sw = ' '.join(tokens_without_sw)
        tokens_without_sw = limpa_texto(tokens_without_sw)
        lista_removida.append(tokens_without_sw)
    
    if type(texto) is str:
        return lista_removida[0]
    elif type(texto) is list:
        return lista_removida
    
```

```python
def busca_idade(texto: str, frase_procurada: str) -> str:
    # extrai a idade em meses. 
    # formatos de entrada esperados:
    # Idade ossea observada de 10 anos = 120 meses.
    # Idade ossea observada entre 13 anos e 6 meses (162 meses) e 14 anos (168 meses).
    
    caracteres = '0123456789,.'
    laudo = texto.split('\n')
    for paragrago in laudo:
        if frase_procurada.upper() in paragrago.upper():
            frase = paragrago.split('=')
            if len(frase) == 1:
                # esta no segundo formato. vou pegar a segunda idade
                frase = paragrago.split('(')
            if len(frase) == 1:
                # sorry, está em algum outro formato
                return np.nan
            # pega o ultimo item da lista
            bloco_idade = frase[-1]
            # pega apenas os caracteres de interesse
            idade = ''.join([x for x in bloco_idade if x in list(caracteres)])
            # caso tenha sobrado algum ponto ou virgula no final, tira
            idade = idade.strip('.')
            idade = idade.strip(',')
            return idade
    return np.nan
```
